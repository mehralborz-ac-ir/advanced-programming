# رسم نمودار با خط و مارکر

```python
x = np.linspace(0, 2*np.pi, 30)
y = np.sin(x)
plt.plot(x, y, '-o')
```

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x, y, marker='+')
```

مارکر	علامت

دایره	“o”

ستاره	“*”

علامت جمع	“+”

مربع	“s”

مثلث	“^”

پنج‌ضلعی	“p”

لیست کامل مارکرها

```python
plt.subplot(2, 2, 1)
```

```python
x = np.linspace(0, 10, 100)
y1 = x
plt.plot(x, y1)
```

```python
plt.subplot(2, 2, 1)
x = np.linspace(0, 10, 100)
y1 = x
plt.plot(x, y1)

plt.subplot(2, 2, 2)
y2 = np.sin(x)
plt.plot(x, y2)

plt.subplot(2, 2, 3)
y3 = x ** 2
plt.plot(x, y3)

plt.subplot(2, 2, 4)
y4 = np.cos(x)
plt.plot(x, y4)
```
